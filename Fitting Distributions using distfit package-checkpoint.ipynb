# Fitting Distributions using distfit package

```python
#importing distfit package in python
import distfit as fit 
```

Read more on the distfit package: https://erdogant.github.io/distfit/pages/html/Abstract.html

```python
#Importing other required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import statsmodels.api as sm 
```

```python
#importing the empirical distribution that we want to fit as dataframe
#pd.read_csv('Enter the file path for your file')
ciapt_data=pd.read_csv('C:/Users/yashs/OneDrive - The University of Manchester/Simulation & Risk Analysis/Distribution Shapes & Fittings/Duration_extracts/CI_Apt_extract.csv')
```

```python
#Cleaning data-Removing Unnnamed column
ciapt_data=ciapt_data.drop('Unnamed: 0',axis=1)
```

```python
ciapt_data.head()
```

```python
#eyeballing the histogram and kernel density
sns.histplot(data=ciapt_data,x='Duration',kde=True)
```

```python
#Fitting all close theoritical distributions on the empirical data
dfit = fit.distfit(distr=["lognorm","weibull_min","gamma","norm","uniform","expon"],alpha=0.05)

#list of all distributions is taken from scipy.stats could be found here:
#https://docs.scipy.org/doc/scipy/reference/stats.html#continuous-distributions

# Fit
dfit.fit_transform(ciapt_data["Duration"])


# Plot seperately
fig, axs = plt.subplots(2,2, figsize=(25,15))
fig, ax = dfit.plot(chart='pdf', n_top=6,ax=axs[0, 0])
fig, ax = dfit.plot(chart='cdf', n_top=6,ax=axs[0, 1])
fig, ax = dfit.qqplot(ciapt_data['Duration'], n_top=6,ax=axs[1, 0])

######p-p plot??
```

```python
#Fitting best distribution plot
dfit = fit.distfit(distr=["lognorm","weibull_min","gamma","norm","uniform","expon"],alpha=0.05)

#list of all distributions is taken from scipy.stats could be found here:
#https://docs.scipy.org/doc/scipy/reference/stats.html#continuous-distributions

# Fit
dfit.fit_transform(ciapt_data["Duration"])


# Plot seperately
fig, axs = plt.subplots(2,2, figsize=(25,15))
fig, ax = dfit.plot(chart='pdf', n_top=1,ax=axs[0, 0])
fig, ax = dfit.plot(chart='cdf', n_top=1,ax=axs[0, 1])
fig, ax = dfit.qqplot(ciapt_data['Duration'], n_top=1,ax=axs[1, 0])


##pp plot????
ln=np.random.lognormal(sigma=0.038,mean=1.942,size=208)
ln
pp_x = sm.ProbPlot(ln, fit=True)
pp_x.ppplot(line="45",marker='o', markerfacecolor='pink', markersize=4,markeredgecolor='red',ax=axs[1,1])
axs[1,1].grid(True, color='gray', linestyle='-', linewidth=0.5)
plt.title("PP Plot")
plt.show()
```

```python
shape=0.59622/0.172231 #mean(i.e. loc)=scale*shape
gam=np.random.gamma(shape, scale=0.172231, size=208)
gam
pp_x = sm.ProbPlot(gam, fit=True)
pp_x.ppplot(line="45",marker='o', markerfacecolor='pink', markersize=4,markeredgecolor='red')
axs[1,1].grid(True, color='gray', linestyle='-', linewidth=0.5)
plt.title("PP Plot")
plt.show()
```
